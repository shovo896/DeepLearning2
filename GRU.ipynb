```python
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler  
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import GRU,Dense 
from tensorflow.keras.optimizers import Adam
```

```python
df=pd.read_csv('data (2).csv',parse_dates=['Date'],index_col='Date')
print(df.head())
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
scaled_data=scaler.fit_transform(df.values)
df
```

data preparation for GRU 

```python
def create_object(data,time_step=1):
    X,y=[],[]
    for i in range(len(data)-time_step-1):
        X.append(data[i:(i+time_step),0])
        y.append(data[i+time_step,0])
    return np.array(X),np.array(y)


time_step=100 
X,y=create_object(scaled_data,time_step)
X=X.reshape(X.shape[0],X.shape[1],1)

```

buiding the GRU model 

```python
model=Sequential()
model.add(GRU(units=50,return_sequences=True,input_shape=(X.shape[1],1)))
model.add(GRU(units=50))
model.add(Dense(units=1))
model.compile(optimizer=Adam(learning_rate=0.001),loss='mean_squared_error')
```

```python
model.fit(X,y,epochs=10,batch_size=32)
```

```python

input_sequence=scaled_data[-time_step:].reshape(1,time_step,1)
predicted_sequence=model.predict(input_sequence)
```

```python
predicted_values=scaler.inverse_transform(predicted_sequence)
print(f"The predicted temperaturte of tghe next day:{predicted_sequence[0][0]:.2f}")
```
