```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
import numpy as np
import matplotlib.pyplot as plt 
```

```python
import cv2
img = cv2.imread("image.jpg")
cv2.imshow("My Image", img)
cv2.waitKey(0)
```

```python
# Example: read an image from disk, display, preprocess and predict with MobileNetV2
import matplotlib.pyplot as plt
# Replace this with your local image path (or use a notebook file picker)
img_path = 'elephant.jpg'  # <-- change to your image file path
# load image and resize to MobileNetV2 expected 224x224
img = image.load_img(img_path, target_size=(224, 224))
# display the image in the notebook
plt.imshow(img)
plt.axis('off')
# convert to array and prepare batch dim
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
# apply MobileNetV2 preprocessing (scaling, etc.)
x = preprocess_input(x)
# load model (uses ImageNet weights) - do this once ideally
model = MobileNetV2(weights='imagenet')
# predict and decode top-3 labels
preds = model.predict(x)
for pred in decode_predictions(preds, top=3)[0]:
    print(f'{pred[1]}: {pred[2]*100:.2f}%')
```
