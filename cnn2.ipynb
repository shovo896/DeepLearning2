```python
base_dir = r"C:\Users\Shuvo\.cache\kagglehub\datasets\karakaggle\kaggle-cat-vs-dog-dataset"

cat_dir = os.path.join(base_dir, 'cats')
dog_dir = os.path.join(base_dir, 'dogs')

print("Cat directory:", os.path.exists(cat_dir))
print("Dog directory:", os.path.exists(dog_dir))
cat_names = os.listdir(cat_dir)
dog_names = os.listdir(dog_dir)
pic_index = 16  # small number to avoid index error
cat_images = [os.path.join(cat_dir, fname) for fname in cat_names[:pic_index//2]]
dog_images = [os.path.join(dog_dir, fname) for fname in dog_names[:pic_index//2]]
fig = plt.gcf()
fig.set_size_inches(16, 16)

for i, img_path in enumerate(cat_images + dog_images):
    sp = plt.subplot(4, 4, i + 1)
    sp.axis('Off')
    img = mpimg.imread(img_path)
    plt.imshow(img)

plt.show()
```
