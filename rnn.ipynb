```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
text="New break up loading!"
chars=sorted(list(text))
char_to_index={char:i for i ,char in enumerate(chars)}
index_to_index={i: char for i ,char in enumerate(chars)}
```

creating sequence and lengths

```python
seq_length=3 
sequences=[]
labels = []
for i  in range(len(text)-seq_length):
       seq=text[i:i+seq_length]
       label=text[i+seq_length]
       sequences.append([char_to_index[char] for char in seq])
       labels.append(char_to_index[label])
x=np.array(sequences)
y=np.array(labels)
```

```python
x_one_hot=tf.one_hot(x,len(chars))
y_one_hot=tf.one_hot(y,len(chars))
```

```python
model=Sequential()
model.add(SimpleRNN(50,input_shape=(seq_length,len(chars)),activation='relu'))
model.add(Dense(len(chars),activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_one_hot,y_one_hot,epochs=100)
```

```python
# Ensure UNK + space support in characters list
if "<UNK>" not in chars:
    chars.append("<UNK>")
if " " not in chars:
    chars.append(" ")

# Update dictionaries
char_to_index = {c: i for i, c in enumerate(chars)}
index_to_char = {i: c for i, c in enumerate(chars)}

UNK = char_to_index["<UNK>"]

start_seq = 'l a d i h r'
seq_length = len(start_seq)
generated_text = start_seq

for i in range(50):
    seq = generated_text[-seq_length:]

    # Encode characters with fallback UNK
    x = np.array([[char_to_index.get(ch, UNK) for ch in seq]])

    # One-hot encoding
    x_one_hot = tf.one_hot(x, len(chars))

    # Predict next char
    prediction = model.predict(x_one_hot, verbose=0)
    next_index = np.argmax(prediction[0])
    next_char = index_to_char.get(next_index, "<UNK>")

    generated_text += next_char

print("Generated Text:")
print(generated_text)
```
